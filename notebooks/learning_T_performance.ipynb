```python
import os
import numpy as np
import pickle
import matplotlib.pyplot as plt
```

```python
names = []
dicts = []
for f in sorted(os.listdir("pickles")):
    filepath = "pickles/" + f + ""
    with open(filepath, "rb") as _f:
        names.append(f)
        dicts.append(pickle.load(_f))
```

```python
plt.figure()
for name, dic in zip(names, dicts):
    plt.plot(dic['T_probabilities'], label=name)
plt.title("Mean Total Probability")
#plt.ylim(0,1.1)
#plt.xlim(-10,250)
plt.legend()
plt.show()
```

```python
plt.figure()
for name, dic in zip(names, dicts):
    plt.plot(dic['T_error'], label=name)
plt.title("Mean Error")
plt.legend()
#plt.xlim(-10,250)
plt.show()
```
